```python
#by user difined function
y=int(input('year'))
m=int(input('month'))
d=int(input('day'))
dic={1:31,2:29,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
count=0
while True:
    if ((y%400==0) or (y%100!=0) and (y%4==0)):
        dic[2]=29
    else:
        dic[2]=28
    dts=dt(y,m,d)
    print(dts)
    
    if m in dic.keys():
        if dts.day==dic[m]:
            if dts.month==12:
                if dts.day==dic[m]:
                    print('kkkkkkkk')
                    m=0
                    y+=1
            m=m+1
            d=0
            
    d+=1
    count+=1
    if count==365:
        break
    
```
